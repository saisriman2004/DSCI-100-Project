### Step 4
#### Method and Plan

For the above I intend to use the KNN (K-Nearest Neighbors) Classificaiton model to predict wheter a player will subscribe to the newsletter based on the age and the number of hours spent playing the game and return and response of `True` or `False` depending upon the prediction KNN is an appropriate model because its a simple and intiutive classification algorithm.It predicts the class of a new observation by looking at the K nearest neighbors in the dataset.

The main idea in KNN algorithm is that similar players behave similarly meaning that players with roughly same age and same number of hours spent playing the game behave similarly when it come to subscribing for the newsletter.One limitation is that the time playes is very large compared to the age so we need to standardize the values.

The dataset will be divided into training and testing dataset. A split of 75% training and 25% testing dataset is ideal to evaluate the model's performance. The models's accuracy will be tested using the cross-validation to find the best values of k.The Evaluation metrics such as the accuracy,prescision,recall and the confusion matric will be used to show how the models classifies subscribes and non-subscribers.The recall will be useful to measure how well the model classifies the players how actually subscribes that is the true postives.
